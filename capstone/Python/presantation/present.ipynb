```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
unoptimized_data = pd.read_csv('C:\\capstone\\Python\\\presantation\\unoptimized.csv')
optimized_data = pd.read_csv('C:\\capstone\\Python\\presantation\\optimized.csv')

# Example metrics to compare
metrics = ['EnergyConsumption', 'Cost', 'IndoorTemp']

# Plot comparison graphs
plt.figure(figsize=(15, 5))

for i, metric in enumerate(metrics):
    plt.subplot(1, len(metrics), i + 1)
    plt.plot(unoptimized_data['Time'], unoptimized_data[metric], label='Unoptimized', color='red')
    plt.plot(optimized_data['Time'], optimized_data[metric], label='Optimized', color='green')
    plt.xlabel('Time')
    plt.ylabel(metric)
    plt.title(f'{metric} Comparison')
    plt.legend()

plt.tight_layout()
plt.show()

# Example analysis - Mean and variance comparison
comparison_results = {}

for metric in metrics:
    unoptimized_mean = unoptimized_data[metric].mean()
    optimized_mean = optimized_data[metric].mean()
    unoptimized_var = unoptimized_data[metric].var()
    optimized_var = optimized_data[metric].var()
    
    comparison_results[metric] = {
        'Unoptimized Mean': unoptimized_mean,
        'Optimized Mean': optimized_mean,
        'Unoptimized Variance': unoptimized_var,
        'Optimized Variance': optimized_var
    }

# Print comparison results
for metric, results in comparison_results.items():
    print(f'{metric} Comparison:')
    for key, value in results.items():
        print(f'{key}: {value}')
    print('\n')

# Example of statistical test - T-test for significance
from scipy.stats import ttest_ind

ttest_results = {}

for metric in metrics:
    t_stat, p_value = ttest_ind(unoptimized_data[metric], optimized_data[metric])
    ttest_results[metric] = {'T-statistic': t_stat, 'P-value': p_value}

# Print T-test results
for metric, results in ttest_results.items():
    print(f'{metric} T-Test:')
    for key, value in results.items():
        print(f'{key}: {value}')
    print('\n')
```
